```python
item_id = "25ef87b7f8744e3e8e8520fe13fb4f9b"
password = "Bracell@21"
login = "Qualidade_Florestal"
portal = "https://gissp.bracell.com/portal"

gis = GIS(url=portal, username=login, password=password)

feature_service = gis.content.get(item_id).layers[0]
url = feature_service.url

for i in range(10001, 90000):
    print(f"Executando consulta para CD_USO_SOLO {i}")
    query = f"CD_USO_SOLO = {i}"
    campos = "MUNICIPIO, ID_PROJETO,ID_TALHAO"

    # Consulta para obter os centróides
    resultados = feature_service.query(where=query, out_fields=campos, return_centroid=True, outSR=4326)
    centroids = resultados.features

    # Exportar centróides para um arquivo CSV
    output_csv = r"F:\Qualidade_Florestal\03- ADMINISTRATIVO\2023\06- COLABORADORES\Gabriel\1 - Processamento de dados\centroides.csv"
    # Escrever os centróides no arquivo CSV
    with open(output_csv, 'a') as f:  # Use 'a' para abrir o arquivo em modo de anexação
        if i == 10001:
            f.write(','.join(centroids[0].attributes.keys()) + '\n')  # Escrever cabeçalho apenas na primeira iteração
        for centroid in centroids:
            f.write(','.join(str(value) for value in centroid.attributes.values()) + '\n')  # Escrever valores

print("Exportação concluída com sucesso!")
```
