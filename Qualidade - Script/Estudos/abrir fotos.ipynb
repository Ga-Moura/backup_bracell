```python
import imageio
from PIL import Image
import numpy as np

def converter_video_para_gif(input_file, output_file, fps=10):
    try:
        # L o vdeo usando o imageio
        video = imageio.get_reader(input_file)
        
        # Obtm as dimenses do vdeo
        width, height = video.get_meta_data()['size']
        
        # Configura o arquivo de sada e o nmero de quadros por segundo (fps)
        writer = imageio.get_writer(output_file, fps=fps)
        
        # Itera atravs de cada quadro do vdeo e adiciona ao arquivo GIF
        for frame in video:
            # Redimensiona o quadro para o tamanho do GIF usando PIL
            frame = Image.fromarray(frame)
            frame = frame.resize((width, height))
            
            # Converte o quadro redimensionado para um array NumPy
            frame_array = np.array(frame)
            
            # Adiciona o quadro ao arquivo GIF
            writer.append_data(frame_array)
        
        # Fecha o arquivo de sada
        writer.close()
        print("Converso concluda com sucesso!")
    
    except Exception as e:
        print(f"Erro ao converter o vdeo para GIF: {e}")

# Exemplo de uso
input_file = r"F:\Qualidade_Florestal\03- ADMINISTRATIVO\2023\06- COLABORADORES\Gabriel\daniel\mapa alfinetes.mp4"  # Substitua pelo caminho do seu arquivo de vdeo
output_file = 'output.gif'      # Especifique o nome do arquivo GIF de sada
converter_video_para_gif(input_file, output_file)
```
