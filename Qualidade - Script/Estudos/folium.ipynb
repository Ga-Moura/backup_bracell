```python
mapa = r"F:\Qualidade_Florestal\03- ADMINISTRATIVO\2023\06- COLABORADORES\Gabriel\1 - Processamento de dados\talhoes_final.json"
```

```python
import geopandas as gpd
import folium

# Carregue o seu arquivo TopoJSON como um GeoDataFrame
gdf = gpd.read_file(mapa)

# Defina um CRS, por exemplo, o WGS 84 (EPSG:4326)
gdf.crs = 'EPSG:4326'

# Crie um mapa
maps = folium.Map(location=[-22.6, -49], zoom_start=10)  # Ajuste a localização e o nível de zoom conforme necessário

# Adicione os dados GeoJSON ao mapa
folium.GeoJson(gdf).add_to(maps)
```
