```python
import pyautogui as py
import time
import pyperclip
```

```python
path =r"https://eflorestal-my.sharepoint.com/:f:/g/personal/resultados_eflorestal_onmicrosoft_com/EsKFrQ07MoZEqMoW8iAh8vYBsc8rlIFXkG3evAtxGJ21Cg?e=OXeBDH"
save = r"F:\Qualidade_Florestal\03- ADMINISTRATIVO\2023\06- COLABORADORES\Gabriel\1 - Processamento de dados"
name = "Controle diário de Produção e Programação"
```

```python
alerta = py.confirm(text ='O seu computador irá realizar uma atualização automática', title ='ALERTA!', buttons = ['Ok, prosseguir', 'Cancelar'])
if alerta == 'Ok, prosseguir':
    #abrindo nova guia
    py.hotkey("ctrlleft", "n")
    time.sleep(2)
    #copiando o link
    pyperclip.copy(path)
    #colando o link
    py.hotkey("ctrl", "v")
    time.sleep(2)
    #apentando enter
    py.press("enter")
    time.sleep(5)
    #clicando na pasta
    py.click(x=308, y=362)
    time.sleep(4)
    #Escolhendo a base para baixar
    py.click(x= 396, y=533 )
    time.sleep(3)
    #clicando em baixar
    py.click(x=459 , y =622)
    time.sleep(6)
    #clicando em abrir a base após baixar
    py.click(x=105 , y =1012)
    time.sleep(5)
    #habilitando edição na base
    py.click(x=1174, y=77)
    time.sleep(4)
    #salvando como utilizando f12
    py.press('F12')
    time.sleep(4)
    #solucionando o erro
    py.press('esc')
    time.sleep(3)
    #selecionando a pasta destino
    py.click(x=580 , y=49 )
    time.sleep(3)
    #apagando o caminho já existente
    py.press('del')
    time.sleep(2)
    #escrevendo novo caminho
    pyperclip.copy(save)
    py.hotkey('ctrl', 'v')
    time.sleep(3)
    #enter para abrir a pasta onde será salvo o arquivo
    py.press('enter')
    time.sleep(2)
    #Clicando para alterar o tipo do arquivo (binário para xlsx)
    py.click(x=515,y=425)
    time.sleep(3)
    #Clicando para escolher o tipo do arquivo
    py.click(x= 487, y=447)
    time.sleep(2)
    #clicando no arquivo para renomeá-lo
    py.click(x=513,y=400)
    time.sleep(2)
    #escrevendo novo nome
    pyperclip.copy(name)
    py.hotkey('ctrl', 'v')
    time.sleep(3)
    #clicando em ok para alterar
    py.press('enter')
    time.sleep(2)
    #clicando em sim para salvar
    py.hotkey('left','left')
    time.sleep(3)
    #clicando em ok para salvar
    py.press('enter')
    time.sleep(4)
    #fechando o excel
    py.hotkey('alt','f4')
    time.sleep(3)
    #redirecionando aba de trabalho
    py.hotkey('win','d')
    
    py.alert(text = 'A atualização automática finalizou', title ='ALERTA' , button = 'Fechar')            
else:
    py.alert(text = 'A atualização automática foi cancelada', title ='ALERTA' , button = 'Fechar')
```
