```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sea
import pandas_profiling as pdp
import openpyxl as plx
```

```python
excel_preparo=r"F:\Qualidade_Florestal\01- SÃO PAULO\02- Silvicultura e Sobrevivência\00 - Arquivos Surveys\Qualidade_de_Preparo_de_Solo.xlsx"
```

```python
names_sheet = plx.load_workbook(excel_preparo)
names_sheet.get_sheet_names()
```

```python
preparo = pd.read_excel(excel_preparo,sheet_name='Qualidade_de_Preparo_de_Solo')
```

```python
coleta =pd.read_excel(excel_preparo, sheet_name="coleta")
```

```python
preparo['objectid'] = preparo['objectid'].astype(str)
```

```python
espacamentos,ax = plt.subplots(1,2,figsize =(20,10))


#espaçamento_entre_linhas
ax[0].hist(preparo['espacamento_entrelinha'],bins=20)
ax[0].set_title("Espaçamento Entre Linhas")
ax[0].set_xlabel("Esp Linhas")
ax[0].set_ylabel("Frequência")


#espaçamento_entre_covas
ax[1].hist(preparo['espacamento_linha'],bins=20)
ax[1].set_title("Espaçamento Entre Covas")
ax[1].set_xlabel("Esp Covas")
ax[1].set_ylabel('Frequência')


plt.show()
```

```python
stand,ax = plt.subplots(1,2,figsize=(20,10))

#stand recomendado
ax[0].hist(preparo['stand_recomendado'],bins=20)
ax[0].set_title("Stand Recomendado")
ax[0].set_xlabel("Stand")
ax[0].set_ylabel('Frequência')

#stand apontado

ax[1].hist(preparo['stand_realizado'],bins=20)
ax[1].set_title("Stand Realizado")
ax[1].set_xlabel("Stand")
ax[1].set_ylabel("Frequência")
plt.show()
```

```python
apontamentos_prof = coleta[['profundidade1','profundidade2','profundidade3','profundidade4','profundidade5','profundidade6','profundidade7','profundidade8','profundidade9','profundidade10']]
# Criar figura e eixos
prof, ax = plt.subplots(2, 5,figsize = (20,10))
# Iterar pelas colunas de 'apontamentos_prof' e plotar os histogramas
for i, row in enumerate(apontamentos_prof.columns):

    linha = i //5  # Ajuste para 5 colunas na matriz de subplots
    coluna = i % 5   # Ajuste para 5 colunas na matriz de subplots

    ax[linha, coluna].hist(apontamentos_prof[row], bins=20)
    ax[linha, coluna].set_title(row.capitalize())  # Capitalizar o nome da coluna
    ax[linha, coluna].set_ylabel('Frequência')
    ax[linha, coluna].set_xlabel("Apontamentos")

plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
linha = coleta[['linha_1','linha_2','linha_3','linha_4','linha_5','linha_6','linha_7','linha_8','linha_9','linha_10']]

apt_linha,ax = plt.subplots(2,5,figsize=(20,10))
for i, row in enumerate(linha.columns):
    linhas_dx = i//5
    coluna = i % 5
    ax[linhas_dx,coluna].hist(linha[row],bins=20)
    ax[linhas_dx,coluna].set_title(row.capitalize())
    ax[linhas_dx,coluna].set_ylabel("Frequência")
    ax[linhas_dx,coluna].set_xlabel("Apontamentos")


plt.tight_layout()

plt.show()
```

```python
covas = coleta[['qtd_covas','cov_1','cov_2','cov_3','cov_4','cov_5','cov_6','cov_7','cov_8','cov_9','cov_10','cov_11','cov_12','cov_13','cov_14','cov_15','cov_16','cov_17','cov_18','cov_19','cov_20']]

apt_covas,ax = plt.subplots(3,7,figsize =(20,10))

for i, row in enumerate(covas.columns):
    linha = i//7
    coluna = i%7
    ax[linha, coluna].hist(covas[row],bins=50)
    ax[linha, coluna].set_title(row.capitalize())
    ax[linha, coluna].set_ylabel("Frequência")
    ax[linha, coluna].set_xlabel("Apontamentos")

plt.tight_layout()
plt.show()
```
