```python
import pandas as pd
import numpy as np
```

```python
path_rf = 'volumes - equipes (gers_vemflors).xlsx'
```

```python
path_fz_regional = r'F:\Silvicultura\01. SP\PROCESSOS\Arquivos Compartilhados - entre Regionais\Fazendas por Regional\Fazendas por Regional_Out2024.xlsx'
```

```python
path_lista_tecnica = 'l - lista_tecnica.xlsx'
```

```python
fzd_regional = pd.read_excel(path_rf,sheet_name= 'Fazenda por eps')
```

```python
#RF_8


rf_8 = pd.read_excel(path_rf, sheet_name='RF08',dtype={'Centro':'object'})

rf_8 = rf_8[pd.notna(rf_8['Centro'])]

rf_8['RF'] = 'RF_8' 
```

```python
#RF_9

rf_9 = pd.read_excel(path_rf, sheet_name='RF09',dtype={'Centro':'object'})

rf_9 = rf_9[pd.notna(rf_9['Centro'])]

rf_9['RF'] = 'RF_9' 
```

```python
#RF_10

rf_10 = pd.read_excel(path_rf, sheet_name='RF10',dtype={'Centro':'object'})

rf_10 = rf_10[pd.notna(rf_10['Centro'])]

rf_10['RF'] = 'RF_10' 
```

```python
rf = pd.concat([rf_8,
               rf_9,
               rf_10])
```

```python
rf['Área planejada - ha'] = pd.to_numeric(rf['Área planejada - ha'], errors='coerce')
```

```python
#Fazenda por regional

fzd_regional = pd.read_excel(path_fz_regional, sheet_name='BASE FAZENDAS TALHÕES TECNICOS',skiprows=1,dtype={'Id Projeto':'object'})

fzd_regional = fzd_regional[['Id Projeto', 'Projeto','EPS DE PLANTIO']]

fzd_regional = fzd_regional[fzd_regional['EPS DE PLANTIO'] !='Sem informação']

fzd_regional = fzd_regional.drop_duplicates('Id Projeto',keep='first')
```

```python
#Regional para RF

rf = rf.merge(fzd_regional, left_on=['Centro'], right_on=['Id Projeto'], how='left')
```

```python
#Lista Técnica

lista_tecnica = pd.read_excel(path_lista_tecnica)

lista_tecnica['status'] = 'Ativo'

lista_tecnica = lista_tecnica[['Operation Description','status']]

lista_tecnica = lista_tecnica.drop_duplicates('Operation Description',keep='first')
```

```python
rf_jfi = rf[(rf['EPS'] == 'JFI') & (pd.notna(rf['SUPERVISOR']))]

rf_jfi = rf_jfi[['SUPERVISOR','EPS']]


rf_jfi = rf_jfi.drop_duplicates('SUPERVISOR', keep='first')



jfi_sup = {'Karine Madruga':'JFI DUARTINA',

'Gustavo Bertoncini':'JFI DUARTINA',

'Emerson Luiz':'JFI DUARTINA',

'Danilo Bavaroski':'JFI CAPÃO',

'André Teixeira':'JFI DOURADO',

'Luciano Bueno':'JFI ITATINGA',

'Pedro Bridi':'JFI DUARTINA',

'DANILO BAVAROSKI':'JFI CAPÃO',

'ANDRÉ TEIXEIRA':'JFI DOURADO',

'Guilherme Inoue':'JFI DUARTINA',

}

rf_jfi['EPS_n'] = rf_jfi['SUPERVISOR'].map(jfi_sup)

```

```python
#Grupo de operações

path_grupo_atividades = 'grupo_atividade.xlsx'

grp_atd = pd.read_excel(path_grupo_atividades)

grp_atd = grp_atd.drop_duplicates(['Grupo de atividade','atividade'], keep='first')

grp_atd = grp_atd.merge(lista_tecnica, left_on=['atividade'], right_on=['Operation Description'], how='left')

grp_atd = grp_atd[grp_atd['status'] == 'Ativo']


grp_atd
```

```python
#Alterando a coluna utilizando o map

rf = rf.merge(rf_jfi[['SUPERVISOR','EPS_n']], left_on='SUPERVISOR', right_on='SUPERVISOR', how='left')


rf['EPS'] = np.where((pd.notna(rf['EPS_n'])) & (rf['EPS'] == 'JFI'), rf['EPS_n'], rf['EPS'])
```

```python
#RF agrupado

rf_groupby = rf.groupby(['Grupo','Atividade','EPS','EPS DE PLANTIO','RF'])['Área planejada - ha'].sum().reset_index()

rf_groupby = rf_groupby.rename({'EPS': 'EPS Planejada',
                                'EPS DE PLANTIO':'EPS - CADASTRO'},axis=1)
```

```python
#Lista técnica

rf_groupby = rf_groupby.merge(lista_tecnica, left_on=['Atividade'], right_on=['Operation Description'], how='left')

rf_groupby = rf_groupby.sort_values('RF', ascending=True)
```
