```python
# Função para verificar se uma coluna é um objeto e pode ser convertida para numérico
def is_convertible_to_numeric(series):
    if series.dtype == 'O':  # Verificar se a coluna é do tipo objeto
        series = series.str.replace(',', '.')  # Substituir vírgulas por pontos
        converted = pd.to_numeric(series, errors='coerce')
        non_null_proportion = converted.notnull().mean()
        return non_null_proportion > 0.2  # Ajustar este limiar conforme necessário
    return False

# Supondo que df_sgf_web e df_sgf são os DataFrames iniciais
convertible_columns_df_sgf_web = [col for col in df_sgf_web.columns if is_convertible_to_numeric(df_sgf_web[col])]
convertible_columns_df_sgf = [col for col in df_sgf.columns if is_convertible_to_numeric(df_sgf[col])]

# Converter as colunas identificadas para numérico
for col in convertible_columns_df_sgf_web:
    df_sgf_web[col] = pd.to_numeric(df_sgf_web[col].str.replace(',', '.'), errors='coerce')

for col in convertible_columns_df_sgf:
    df_sgf[col] = pd.to_numeric(df_sgf[col].str.replace(',', '.'), errors='coerce')

print("Colunas convertíveis para numérico em df_sgf_web:", convertible_columns_df_sgf_web)
print("Colunas convertíveis para numérico em df_sgf:", convertible_columns_df_sgf)
```

```python
# Função para padronizar colunas
def standardize_column(col, convertible_columns):
    # Remover espaços em branco extras se a coluna for de texto
    if col.name in convertible_columns:
        return pd.to_numeric(col, errors='coerce')
    elif col.dtype == 'object':
        return col.str.strip()
    else:
        return col

# Padronizando todas as colunas dos DataFrames
df_sgf_web = df_sgf_web.apply(lambda col: standardize_column(col, convertible_columns_df_sgf_web))
df_sgf = df_sgf.apply(lambda col: standardize_column(col, convertible_columns_df_sgf))
```

```python
# Função para comparar duas colunas
def compare_columns(col1, col2):
    # Comparar apenas os valores válidos (não nulos) e eliminar espaços em branco
    col1_valid = col1.dropna().astype(str).str.strip()
    col2_valid = col2.dropna().astype(str).str.strip()
    # Verificar se há uma correspondência significativa entre as colunas
    return col1_valid.equals(col2_valid)

# Criação do dicionário de mapeamento
column_mapping = {}
lista_boletim_full = list(set(df_sgf_web["Id"]) & set(df_sgf["DCR_COD_BOLETIM"]))
#lista_boletim_full =[465638,465070,489400,465564,463305,465629,474329,477617,481134,461947]
# Iteração sobre todas as combinações de colunas de df1 e df2
for col1 in df_sgf.columns:
    print(col1)
    for col2 in df_sgf_web.columns:
        if compare_columns( df_sgf[df_sgf['DCR_COD_BOLETIM'].isin(lista_boletim_full)].reset_index()[col1],df_sgf_web[df_sgf_web['Id'].isin(lista_boletim_full)].reset_index()[col2]):
            if col1 in column_mapping:
                column_mapping[col1].append(col2)
            else:
                column_mapping[col1] = [col2]

# Exibindo o dicionário de mapeamento
print(column_mapping)
```
